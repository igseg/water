```python
class dam_data:
    def __init__(self, cap_max = 0, cap_min = 0, dam_height = 0, area_skm = 0, river = 0, near_city = 0, dam_name = None, main_use = None, state = None):
        self.cap_max    = cap_max
        self.cap_min    = cap_min
        self.dam_height = dam_height
        self.area_skm   = area_skm
        self.river      = river
        self.near_city  = near_city
        self.name       = dam_name
        self.main_use   = main_use
        self.state      = state
        
    def __str__(self):
           return f'id is: {self.id}\n name is: {self.name}\n main use is: {self.main_use}\n maximum capacity is: {self.cap_max}\n minimum capacity is: {self.cap_min}\n height of the dam is: {self.dam_height}\n area is: {self.area_skm}\n river is: {self.river}\n nearest city is: {self.near_city}'
```

```python
ids =  get_time_series_ids()

information = []
i = 0
missing = 0
for dam_id in ids:
    

    dam = from_id_to_dam_data(dam_id)
    
    if dam == None:
        missing += 1
        continue
    if dam.state != 'California':
        continue
    if dam.main_use != 'Hydroelectric':
        continue

    corr, nan_data, num_rows = id_to_series_data(dam_id)
    
    information.append([dam,corr,nan_data,num_rows])
    i+=1
    
## Warnings are caused by dams that have all inflows and/or outflows missings, making impossible correlation's calculus.
## We delete those that have NaN in inflow and/or outflow

information_withoutnan = []


counter = 0
for item in information:
    if np.isnan(item[1]):
        print(item[2])
        counter += 1
        continue
    information_withoutnan.append(item)
    
print(counter)
```

```python
starting_date = datetime(2004,1,1)
ending_date   = datetime(2021,12,31)

missing_ratios = []

for item in information_withoutnan:
    dam_id  = item[0].id
    path    = from_id_to_time_series(dam_id)
    ts      = pd.read_csv(path)
    print(ts.date.values[-1])
    print('='*70)
    ts.date = ts.date.apply(lambda x: datetime(int(x[:4]),
                                     int(x[5:7]),
                                     int(x[8:10])
    ))

    ts = ts[(ts['date']> starting_date ) & (ts['date']< ending_date)]

    ratio_missings = np.isnan(ts).sum(axis = 0)/ts.shape[0] * 100
    missing_ratios.append((ratio_missings, ts.shape[0] * 100, dam_id ))
    
```
